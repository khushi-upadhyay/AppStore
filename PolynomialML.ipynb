```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv("/content/drive/MyDrive/ML/temperature_vs_icecream.csv")
```

```python
df.head()
```

```python
X = df['Temperature (°C)']
y = df['Ice Cream Sales']
```

```python
X
```

```python
y
```

```python
plt.scatter(X, y)
plt.show()
```

```python
lmodel = LinearRegression()
lmodel.fit(X.values.reshape(-1,1), y)
```

```python
ypred = lmodel.predict(X.values.reshape(-1,1))
```

```python
plt.scatter(X, y)
plt.plot(X, ypred, color='red')
plt.show()
```

```python
poly1 = PolynomialFeatures(degree=2)
X_poly = poly1.fit_transform(X.values.reshape(-1,1))
```

```python
poly1.fit(X_poly, y)
```

```python
PolyModel = LinearRegression()
PolyModel.fit(X_poly, y)
predPoly = PolyModel.predict(X_poly)
```

```python
plt.scatter(X, y)
plt.plot(X, predPoly, color='green')
plt.show()
```

```python
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(X.values.reshape(-1,1))
```

```python
poly3.fit(X_poly3, y)
```

```python
PolyModel3 = LinearRegression()
PolyModel3.fit(X_poly3, y)
predPoly3 = PolyModel3.predict(X_poly3)
plt.scatter(X, y)
plt.plot(X, predPoly3, color='orange')
plt.show()

```

```python
poly4 = PolynomialFeatures(degree=4)
X_poly4 = poly4.fit_transform(X.values.reshape(-1,1))
```

```python
poly4.fit(X_poly4, y)
```

```python
PolyModel4 = LinearRegression()
PolyModel4.fit(X_poly4, y)
predPoly4 = PolyModel4.predict(X_poly4)
plt.scatter(X, y)
plt.plot(X, predPoly4, color='orange')
plt.show()
```

```python
# prompt: do for degree 5

poly5 = PolynomialFeatures(degree=5)
X_poly5 = poly5.fit_transform(X.values.reshape(-1,1))

poly5.fit(X_poly5, y)

PolyModel5 = LinearRegression()
PolyModel5.fit(X_poly5, y)
predPoly5 = PolyModel5.predict(X_poly5)
plt.scatter(X, y)
plt.plot(X, predPoly5, color='purple')
plt.show()
```

```python

poly8 = PolynomialFeatures(degree=8)
X_poly8 = poly8.fit_transform(X.values.reshape(-1,1))

poly8.fit(X_poly8, y)

PolyModel8 = LinearRegression()
PolyModel8.fit(X_poly8, y)
predPoly8 = PolyModel8.predict(X_poly8)
plt.scatter(X, y)
plt.plot(X, predPoly8, color='brown')
plt.show()
```

```python

poly10 = PolynomialFeatures(degree=10)
X_poly10 = poly10.fit_transform(X.values.reshape(-1,1))

poly10.fit(X_poly10, y)

PolyModel10 = LinearRegression()
PolyModel10.fit(X_poly10, y)
predPoly10 = PolyModel10.predict(X_poly10)
plt.scatter(X, y)
plt.plot(X, predPoly10, color='black')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse_values = {}
mse_values[1] = mean_squared_error(y, ypred)
mse_values[2] = mean_squared_error(y, predPoly)
mse_values[3] = mean_squared_error(y, predPoly3)
mse_values[4] = mean_squared_error(y, predPoly4)
mse_values[5] = mean_squared_error(y, predPoly5)
mse_values[8] = mean_squared_error(y, predPoly8)
mse_values[10] = mean_squared_error(y, predPoly10)

print("MSE values for different polynomial degrees:")
for degree, mse in mse_values.items():
  print(f"Degree {degree}: {mse}")
```

```python
print(f"Underfitting")
print(f"Degree 1 (MSE = {mse_values[1]})")
print(f"Degree 2 (MSE = {mse_values[2]})")
print(f"Degree 3 (MSE = {mse_values[3]})")
print(f"Degree 4 (MSE = {mse_values[4]})")
print(f"Degree 10 (MSE = {mse_values[10]})")
```
